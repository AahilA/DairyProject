```python
cowfilenames = glob.glob("../DA_9_CD/*.csv")
display(env_data_file)
```

```python
milk_data_file = pd.merge(milk_data_file, env_data_file, left_on=['Date'], right_on=['Date'])
```

```python
milk_data_file = milk_data_file.dropna(axis = 1, how = 'all')
milk_data_file.head(100)
```

```python
cow2frame = {}

for file in cowfilenames:
    try:
#         if(not ("t_1.csv" in file)):
#             continue
        cow_data_file = pd.read_csv(file)
        cowID = cow_data_file.at[0,'CowID']
  #      print(cowID)
        cow_milk_data = milk_data_file[milk_data_file['AnimalId'] == cowID]
        
        f = dict.fromkeys(cow_milk_data, 'mean')
        del f['Date']
        f['Session'] = 'size'
        f['Yield(gr)'] = 'sum'
        f['YieldAverage(gr)'] = 'sum'
        cow_milk_data = cow_milk_data.resample('d', on='Date').agg(f).dropna(how='all')

        cow_data_file['time'] = pd.to_datetime(cow_data_file.time)
        cow_data_file['temp_without_drink_cycles_std'] = cow_data_file['temp_without_drink_cycles']
        #display(cow_data_file['temp_without_drink_cycles_std'])
        f2 = dict.fromkeys(cow_data_file, 'mean')
        del f2['time']
        f2['temp_without_drink_cycles_std'] = 'std'
        
        cow_data_file = cow_data_file.resample('d', on='time').agg(f2).dropna(how='all')
        cow_data_file = pd.merge(cow_data_file, cow_milk_data, left_index=True, right_on=['Date'])
        #print(len(cow_data_file))
        if(len(cow_data_file)>0):
            cow2frame[cowID] = cow_data_file
    except:
        print(file)
    
```

```python
cow2frame[11146]
```

```python
for key in cow2frame:
    for col in cow2frame[key]:
        cow2frame[key]['median']= cow2frame[key][col].rolling(10).median()
        cow2frame[key]['std'] = cow2frame[key][col].rolling(10).std()

        #filter setup
        cow2frame[key] = cow2frame[key][(cow2frame[key][col] <= cow2frame[key]['median']+3*cow2frame[key]['std']) &
                                        (cow2frame[key][col] >= cow2frame[key]['median']-3*cow2frame[key]['std'])]
        
        cow2frame[key].drop(['median', 'std'], axis=1)
```

```python
for key in cow2frame:
    cow2frame[key].to_csv(path_or_buf=('./aggregatedData/cowID'+str(key) + '.csv'))
```
